```python
###Drop Nan Values
df=df.dropna()
```

```python
## Get the Independent Features

X=df.drop('label',axis=1)
```

```python
## Get the Dependent features
y=df['label']
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM,Dense
```

```python
### Vocabulary size just consider dictionary of words u can give any value
voc_size=5000
```

```python
#Onehot Representation
messages=X.copy()
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
#Embedding Representation
sent_length=20 #to make all the sen of equal shape
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## Creating model
from tensorflow.keras.layers import Dropout
embedding_vector_features=40 # in embedding layer it takes input and converts some specific  no of features or vectors u can give any no
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100)) # one layer having 100 neurons
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
### Finally Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
